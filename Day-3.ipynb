### Pandas data structures
* 1. Series
* 2. Data Frame

```python
import pandas as pd

st={'name':['shiva','ganesh','haritha'],
   'rollnum':[12,13,15],
   'branch':['cse','IT','ECE']}

df=pd.DataFrame(st)
df
```

```python
df['marks']=[98,76,56]  # to add a new column
df
```

```python
df.loc[3]=['parvathi',15,'EEE',95]  # to add new row
df
```

```python
# to change the index
df.index=['a','b','c','d']
df
```

```python
# to delete data from dataframe

# drop
# axis=0 row
# axis=1 column
```

```python
df.drop('a',axis=0,inplace=True)
df
```

```python
df.drop('d',axis=0,inplace=True)
df
```

```python
# to delete a column 'marks'

df.drop('marks',axis=1,inplace=True)
df
```

```python
# to delete all columns
df.drop(df.columns,axis=1,inplace=True)
```
